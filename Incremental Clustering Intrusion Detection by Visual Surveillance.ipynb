```python
from sklearn import datasets
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
```

```python
iris = datasets.load_iris()
```

```python
X = iris.data[:, :2]
y = iris.target
```

```python
plt.scatter(X[:,0], X[:,1], c=y, cmap='gist_rainbow')
plt.xlabel('Spea1 Length')
plt.ylabel('Sepal Width')
```

```python
km = KMeans(n_clusters = 3, n_jobs = 4, random_state=21)
km.fit(X)
```

```python
centers = km.cluster_centers_
print(centers)
```

```python
new_labels = km.labels_
fig, axes = plt.subplots(1, 2, figsize=(16,8))
axes[0].scatter(X[:, 0], X[:, 1], c=y, cmap='gist_rainbow',
edgecolor='k', s=150)
axes[1].scatter(X[:, 0], X[:, 1], c=new_labels, cmap='jet',
edgecolor='k', s=150)
axes[0].set_xlabel('Sepal length')
axes[0].set_ylabel('Sepal width')
axes[1].set_xlabel('Sepal length')
axes[1].set_ylabel('Sepal width')
axes[0].set_title('Actual')
axes[1].set_title('Predicted')
```

```python
import cv2
from IPython.display import clear_output
import os
if not os.path.exists('vid/') :
    os.mkdir('vid/')
    cap= cv2.VideoCapture('vid_surv.mp4')
    i=0
    while(cap.isOpened()):
        ret, frame = cap.read()
        if ret == False:
            break
        cv2.imwrite("vid/"+str(i)+'.jpg',frame)
        i+=1

    cap.release()
    cv2.destroyAllWindows()
```

```python
Images = []
import numpy as np
for i in range(len(os.listdir('vid/'))) :
    img = np.ravel(cv2.resize(cv2.imread('vid/'+str(i)+'.jpg',0),(100,100)))
    Images.append(img)
Images = np.array(Images)/255.
```

```python
km = KMeans(n_clusters = 2, n_jobs = 4, random_state=21)
km.fit(Images)
```

```python
centers = km.cluster_centers_
print(centers)
```
